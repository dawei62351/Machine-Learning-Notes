## Problem 2.2

```python
import numpy as np
import matplotlib.pyplot as plt

weights = [164,158,172,153,144,156,189,163,134,159,143,176,177,162,141,151,182,185,171,152]

mean = np.mean(weights)
median = np.median(weights)
standard_deviation = np.std(weights)
print(len(weights))
print(mean)
print(median)
print(standard_deviation)

plt.figure()

plt.hist(np.arange(len(weights))+1, bins=np.arange(0.5, len(weights) + 1.5, 1), weights=weights, edgecolor='black')
bars = plt.bar(np.arange(len(weights)) + 1, weights)
plt.xlabel('people')
plt.ylabel('weights')
plt.title('Weight value for 20 people')
plt.xticks(np.arange(1, len(weights) + 1, 1))
for bar, value in zip(bars, weights):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(value), ha='center', va='bottom')

plt.show
```
